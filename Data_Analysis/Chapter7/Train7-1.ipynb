            <li>丢弃<code>呆账</code>、<code>强制停卡</code>、<code>退票</code>为1且<code>拒往记录</code>为2的记录。</li>
            <li>丢弃<code>频率</code>为5且<code>月刷卡额</code>不为1的数据。</li>

```python
# 读取信用卡数据
import pandas
credit_cards = pandas.read_csv('data/credit_card.csv', encoding='GB18030')
print('信用卡信息数据规模为：', credit_cards.shape)
```

```python
# 丢弃逾期数据为1
overdue = (credit_cards['逾期'] != 1)
# 丢弃呆账数据为1
bad_debt = (credit_cards['呆账'] != 1)
# 丢弃强制停卡为1
force_lock = (credit_cards['强制停卡记录'] != 1)
# 丢弃退票为1
refund = (credit_cards['退票'] != 1)
# 丢弃拒往记录为1
refused_1 = (credit_cards['拒往记录'] != 1)
# 丢弃瑕疵户为2
flaw = (credit_cards['瑕疵户'] != 2)
# 丢弃拒往记录为2
refused_2 = (credit_cards['拒往记录'] != 2)
# 丢弃频率为5
frequency_5 = (credit_cards['频率'] != 5)
# 丢弃月刷卡额不为1
quota = (credit_cards['月刷卡额'] == 1)
```

```python
# 记录组 A
group_A = credit_cards[overdue & bad_debt & force_lock & refund & refused_1 & flaw]
print('第A组记录规模为：', group_A.shape)
# 记录组 B
group_B = credit_cards[bad_debt & force_lock & refund & refused_2]
print('第B组记录规模为：', group_B.shape)
# 记录组 C
group_C = credit_cards[frequency_5 & quota]
print('第C组记录规模为：', group_C.shape)
```
