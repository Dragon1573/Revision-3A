# 任务3.1　掌握绘图基础语法与常用参数

## 代码3-1　`pyplot`中的基础绘图语法

```python
import numpy
from matplotlib import pyplot
# 在Jupyter Notebook中穿插展示pyplot图表
%matplotlib inline
data = numpy.arange(0, 1, 0.01)
# 标题
pyplot.title('lines')
# 坐标轴标签
pyplot.xlabel('x')
pyplot.ylabel('y')
# 坐标轴展示范围
pyplot.xlim((0, 1))
pyplot.ylim((0, 1))
# 坐标轴刻度标签
pyplot.xticks([0, 0.2, 0.4, 0.6, 0.8, 1])
pyplot.yticks([0, 0.2, 0.4, 0.6, 0.8, 1])
# 绘制曲线
pyplot.plot(data, data ** 2)
pyplot.plot(data, data ** 4)
# 图例
pyplot.legend(['y=x^2', 'y=x^4'])
# 图像保存（图标最好使用SVG矢量图保存，支持任意无损缩放）
pyplot.savefig('figures/C3-1-1.svg')
pyplot.show()
```

# 代码3-2　包含子图绘制的基础语法

```python
rad = numpy.arange(0, 2 * numpy.pi, 0.01)
# 第1幅子图
figure = pyplot.figure(figsize=(8, 6), dpi=80)
subplot1 = figure.add_subplot(2, 1, 1)
pyplot.title('lines')
pyplot.xlabel('x')
pyplot.ylabel('y')
pyplot.xlim((0, 1))
pyplot.ylim((0, 1))
pyplot.xticks(numpy.arange(0, 1, 0.2))
pyplot.yticks(numpy.arange(0, 1, 0.2))
pyplot.plot(rad, rad ** 2)
pyplot.plot(rad, rad ** 4)
pyplot.legend(['y=x^2', 'y=x^4'])
# 第2幅子图
subplot2 = figure.add_subplot(2, 1, 2)
pyplot.title('sin/cos')
pyplot.xlabel('rad')
pyplot.ylabel('value')
pyplot.xlim((0, 2 * numpy.pi))
pyplot.ylim((-1, 1))
pyplot.xticks(numpy.arange(0, 2 * numpy.pi, numpy.pi / 2))
pyplot.yticks(numpy.arange(-1, 1, 0.5))
pyplot.plot(rad, numpy.sin(rad))
pyplot.plot(rad, numpy.cos(rad))
pyplot.legend(['sin', 'cos'])
pyplot.savefig('figures/C3-1-2.svg')
pyplot.show()
```

## 代码3-3　调节线条的rc参数

```python
# 原图
x = numpy.linspace(0, 4 * numpy.pi)
y = numpy.sin(x)
pyplot.plot(x, y, label="$sin(x)$")
pyplot.title('sin')
pyplot.savefig('figures/C3-1-3.svg')
pyplot.show()
```

```python
# 修改rc参数后的图
pyplot.rcParams['lines.linestyle'] = '-.'
pyplot.rcParams['lines.linewidth'] = 3
pyplot.plot(x, y, label='$sin(x)$')
pyplot.title('sin')
pyplot.savefig('figures/C3-1-4.svg')
pyplot.show()
```

## 代码3-4　调节字体的rc参数

```python
# 无法显示中文标题
pyplot.plot(x, y, label='$sin(x)$')
pyplot.title('sin曲线')
pyplot.savefig('figures/C3-1-5.svg')
pyplot.show()
```

```python
# 设置rc参数显示中文标题
# 设置字体为微软雅黑显示中文
pyplot.rcParams['font.sans-serif'] = 'Microsoft YaHei'
pyplot.rcParams['axes.unicode_minus'] = False
pyplot.plot(x, y, label='$sin(x)$')
pyplot.title('sin曲线')
pyplot.savefig('figures/C3-1-6.svg')
pyplot.show()
```
