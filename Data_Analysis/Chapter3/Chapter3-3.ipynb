# 任务3.3　分析特征内部数据分布与分散情况

## 代码3-12　绘制2017年第一季度各产业国民生产总值直方图

```python
import numpy
from matplotlib import pyplot
%matplotlib inline
pyplot.rcParams['font.sans-serif'] = "Microsoft YaHei"
pyplot.rcParams['axes.unicode_minus'] = False
data = numpy.load('data/Task3-2.npz', allow_pickle=True)
values = data['values']
label = ['第一产业', '第二产业', '第三产业']
pyplot.figure(figsize=(6, 5))
pyplot.bar(range(3), values[-1, 3:6], width=0.5)
pyplot.xlabel('产业')
pyplot.ylabel('生产总值（亿元人民币）')
pyplot.xticks(range(3), label)
pyplot.title('2017年第一季度各产业国民生产总值直方图')
pyplot.savefig('figures/C3-3-1.svg')
pyplot.show()
```

## 代码3-13　绘制2017年第一季度各产业国民生产总值饼图

```python
pyplot.figure(figsize=(6, 6))
label = ['第一产业', '第二产业', '第三产业']
explode = [0.01, 0.01, 0.01]
pyplot.pie(values[-1, 3:6], explode=explode, labels=label, autopct='%1.2f%%')
pyplot.title('2017年第一季度各产业国民生产总值饼图')
pyplot.savefig('figures/C3-3-2.svg')
pyplot.show()
```

## 代码3-14　绘制$2000\sim2017$年各产业国民生产总值箱线图

```python
label = ['第一产业', '第二产业', '第三产业']
gdp = (list(values[:, 3]), list(values[:, 4]), list(values[:, 5]))
pyplot.figure(figsize=(6, 4))
pyplot.boxplot(gdp, notch=True, labels=label, meanline=True)
pyplot.title('2000~2017年各产业国民生产总值箱线图')
pyplot.savefig('figures/C3-3-3.svg')
pyplot.show()
```

## 代码3-15　绘制国民生产总值构成分布直方图

```python
import numpy
from matplotlib import pyplot
%matplotlib inline
data = numpy.load('data/Task3-2.npz', allow_pickle=True)
values = data['values']
pyplot.rcParams['font.sans-serif'] = "Microsoft YaHei"
pyplot.rcParams['axes.unicode_minus'] = False
label1 = ['第一产业', '第二产业', '第三产业']
label2 = [
    '农业', '工业', '建筑', '批发', '交通',
    '餐饮', '金融', '房地产', '其他'
]
p = pyplot.figure(figsize=(12, 12))
p.add_subplot(2, 2, 1)
pyplot.bar(range(3), values[0, 3:6], width=0.5)
pyplot.xlabel('产业')
pyplot.ylabel('生产总值（亿元人民币）')
pyplot.xticks(range(3), label1)
pyplot.title('2000年第一季度国民生产总值产业构成分布直方图')
p.add_subplot(2, 2, 2)
pyplot.bar(range(3), values[-1, 3:6], width=0.5)
pyplot.xlabel('产业')
pyplot.ylabel('生产总值（亿元人民币）')
pyplot.xticks(range(3), label1)
pyplot.title('2017年第一季度国民生产总值产业构成分布直方图')
p.add_subplot(2, 2, 3)
pyplot.bar(range(9), values[0, 6:], width=0.5)
pyplot.xlabel('行业')
pyplot.ylabel('生产总值（亿元人民币）')
pyplot.xticks(range(9), label2)
pyplot.title('2000年第一季度国民生产总值行业构成分布直方图')
p.add_subplot(2, 2, 4)
pyplot.bar(range(9), values[-1, 6:], width=0.5)
pyplot.xlabel('行业')
pyplot.ylabel('生产总值（亿元人民币）')
pyplot.xticks(range(9), label2)
pyplot.title('2017年第一季度国民生产总值行业构成分布直方图')
pyplot.savefig('figures/C3-3-4.svg')
pyplot.show()
```

## 代码3-16　绘制国民生产总值构成分布饼图

```python
label1 = ['第一产业', '第二产业', '第三产业']
label2 = [
    '农业', '工业', '建筑', '批发', '交通',
    '餐饮', '金融', '房地产', '其他'
]
explode1 = 3 * [0.01]
explode2 = 9 * [0.01]
p = pyplot.figure(figsize=(12, 12))
p.add_subplot(2, 2, 1)
pyplot.pie(values[0, 3:6], explode=explode1, labels=label1, autopct='%1.2f%%')
pyplot.title('2000年第一季度国民生产总值产业构成分布饼图')
p.add_subplot(2, 2, 2)
pyplot.pie(
    values[-1, 3:6], explode=explode1, labels=label1, autopct='%1.2f%%'
)
pyplot.title('2017年第一季度国民生产总值产业构成分布饼图')
p.add_subplot(2, 2, 3)
pyplot.pie(
    values[0, 6:], explode=explode2, labels=label2,
    autopct='%1.1f%%', pctdistance=0.75
)
pyplot.title('2000年第一季度国民生产总值行业构成分布饼图')
p.add_subplot(2, 2, 4)
pyplot.pie(values[-1, 6:], explode=explode2, labels=label2, autopct='%1.2f%%')
pyplot.title('2017年第一季度国民生产总值行业构成分布饼图')
pyplot.savefig('figures/C3-3-5.svg')
pyplot.show()
```

# 代码3-17　绘制国民生产总值分散情况箱线图

```python
label1 = ['第一产业', '第二产业', '第三产业']
label2 = [
    '农业', '工业', '建筑', '批发', '交通',
    '餐饮', '金融', '房地产', '其他'
]
gdp1 = (list(values[:, 3]), list(values[:, 4]), list(values[:, 5]))
gdp2 = ([list(values[:, i]) for i in range(6, 15)])
p = pyplot.figure(figsize=(8, 8))
p.add_subplot(2, 1, 1)
pyplot.boxplot(gdp, notch=True, labels=label1, meanline=True)
pyplot.title('2000~2017年各产业国民生产总值箱线图')
pyplot.ylabel('生产总值（亿元人民币）')
p.add_subplot(2, 1, 2)
pyplot.boxplot(gdp2, notch=True, labels=label2, meanline=True)
pyplot.title('2000~2017年各行业国民生产总值箱线图')
pyplot.xlabel('行业')
pyplot.ylabel('生产总值（亿元人民币）')
pyplot.savefig('figures/C3-3-6.svg')
pyplot.show()
```
