# 操作题

```python
import numpy
from matplotlib import pyplot
%matplotlib inline
# pyplot参数设置
pyplot.rcParams['font.sans-serif'] = "Microsoft YaHei"
pyplot.rcParams['axes.unicode_minus'] = False

from sklearn.datasets import load_iris
# 读取数据集
iris_data = load_iris()
values = iris_data['data']
names = ('花萼长度', '花萼宽度', '花瓣长度', '花瓣宽度')
```

1. 读取鸢尾花数据集，使用循环和子图绘制各个特征之间的散点图

```python
figure_A = pyplot.figure(figsize=(10, 8))
for plot_ID in range(0, 4):
    figure_A.add_subplot(4, 1, plot_ID + 1)
    pyplot.scatter(range(150), values[:, plot_ID])
    pyplot.ylabel('单位：cm')
    pyplot.title('鸢尾' + names[plot_ID] + '散点图')
figure_A.tight_layout()
pyplot.savefig('figures/Q3-1.svg')
pyplot.show()
```

2. 绘制各个特征的箱线图，查看是否存在异常值。

```python
pyplot.figure(figsize=(10, 8))
boxes = [values[:, column] for column in range(4)]
pyplot.boxplot(boxes, notch=True, labels=names, meanline=True)
pyplot.xlabel('特征')
pyplot.ylabel('单位：cm')
pyplot.title('鸢尾花特征分布箱线图')
pyplot.savefig('figures/Q3-2.svg')
pyplot.show()
```
