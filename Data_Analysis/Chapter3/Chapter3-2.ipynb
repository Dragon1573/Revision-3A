# 任务3.2　分析特征间的关系

## 代码3-5　绘制$2000\sim2017$年各季度国民生产总值散点图

```python
import numpy
from matplotlib import pyplot
%matplotlib inline
pyplot.rcParams['font.sans-serif'] = 'Microsoft YaHei'
pyplot.rcParams['axes.unicode_minus'] = False
data = numpy.load('data/Task3-2.npz', allow_pickle=True)
name = data['columns']
values = data['values']
pyplot.figure(figsize=(8, 7))
pyplot.scatter(values[:, 0], values[:, 2], marker='o')
pyplot.xlabel('年份')
pyplot.ylabel('生产总值（亿元人民币）')
pyplot.ylim((0, 225000))
pyplot.xticks(range(0, 70, 4), values[range(0, 70, 4), 1], rotation=45)
pyplot.title('2000~2017年各季度国民生产总值散点图')
pyplot.savefig('figures/C3-2-1.svg')
pyplot.show()
```

## 代码3-6　绘制$2000\sim2017$年间各产业各季度国民生产总值的散点图

```python
pyplot.figure(figsize=(8, 7))
# 图1
pyplot.scatter(values[:, 0], values[:,3], marker='o', c='red')
# 图2
pyplot.scatter(values[:, 0], values[:, 4], marker='D', c='blue')
# 图3
pyplot.scatter(values[:, 0], values[:, 5], marker='v', c='yellow')
pyplot.xlabel('年份')
pyplot.ylabel('生产总值（亿元人民币）')
pyplot.xticks(range(0, 70, 4), values[range(0, 70, 4), 1], rotation=45)
pyplot.title('2000~2017年各产业各季度国民生产总值散点图')
pyplot.legend(['第一产业', '第二产业', '第三产业'])
pyplot.savefig('figures/C3-2-2.svg')
pyplot.show()
```

## 代码3-10　绘制$2000\sim2017$年各产业与行业的国民生产总值散点图

```python
import numpy
from matplotlib import pyplot
%matplotlib inline
pyplot.rcParams['font.sans-serif'] = 'Microsoft YaHei'
pyplot.rcParams['axes.unicode_minus'] = False
data = numpy.load('data/Task3-2.npz', allow_pickle=True)
name = data['columns']
values = data['values']
p = pyplot.figure(figsize=(12, 12))
# 子图A
p.add_subplot(2, 1, 1)
pyplot.scatter(values[:, 0], values[:, 3], marker='o', c='r')
pyplot.scatter(values[:, 0], values[:, 4], marker='D', c='b')
pyplot.scatter(values[:, 0], values[:, 5], marker='v', c='y')
pyplot.ylabel('生产总值（亿元人民币）')
pyplot.title('2000~2017年各产业与行业各季度国民生产总值散点图')
pyplot.legend(['第一产业', '第二产业', '第三产业'])
# 子图B
p.add_subplot(2, 1, 2)
pyplot.scatter(values[:, 0], values[:, 6], marker='o', c='r')
pyplot.scatter(values[:, 0], values[:, 7], marker='D', c='b')
pyplot.scatter(values[:, 0], values[:, 8], marker='v', c='y')
pyplot.scatter(values[:, 0], values[:, 9], marker='8', c='g')
pyplot.scatter(values[:, 0], values[:, 10], marker='p', c='c')
pyplot.scatter(values[:, 0], values[:, 11], marker='+', c='m')
pyplot.scatter(values[:, 0], values[:, 12], marker='s', c='k')
pyplot.scatter(values[:, 0], values[:, 13], marker='*', c='purple')
pyplot.scatter(values[:, 0], values[:, 14], marker='d', c='brown')
pyplot.legend([
    '农业', '工业', '建筑', '批发', '交通', '餐饮', '金融', '房地产', '其他'
])
pyplot.xlabel('年份')
pyplot.ylabel('生产总值（亿元人民币）')
pyplot.xticks(range(0, 70, 4), values[range(0, 70, 4), 1], rotation=45)
pyplot.savefig('figures/C3-2-6.svg')
pyplot.show()
```

## 代码3-11　绘制$2000\sim2017$年各产业与行业的国民生产总值折线图

```python
p1 = pyplot.figure(figsize=(8, 7))
p1.add_subplot(2, 1, 1)
pyplot.plot(
    values[:, 0], values[:, 3], 'b-',
    values[:, 0], values[:, 4], 'r-.',
    values[:, 0], values[:, 5], 'g--'
)
pyplot.ylabel('生产总值（亿元人民币）')
pyplot.title('2000~2017年各产业与行业国民生产总值折线图')
pyplot.legend(['第一产业', '第二产业', '第三产业'])
p1.add_subplot(2, 1, 2)
pyplot.plot(
    values[:, 0], values[:, 6], 'r-',
    values[:, 0], values[:, 7], 'b-.',
    values[:, 0], values[:, 8], 'y--',
    values[:, 0], values[:, 9], 'g:',
    values[:, 0], values[:, 10], 'c-',
    values[:, 0], values[:, 11], 'm-.',
    values[:, 0], values[:, 12], 'k--',
    values[:, 0], values[:, 13], 'r:',
    values[:, 0], values[:, 14], 'b-',
)
pyplot.legend([
    '农业', '工业', '建筑', '批发', '交通', '餐饮', '金融', '房地产', '其他'
])
pyplot.xlabel('年份')
pyplot.ylabel('生产总值（亿元人民币）')
pyplot.xticks(range(0, 70, 4), values[range(0, 70, 4), 1], rotation=45)
pyplot.savefig('figures/C3-2-7.svg')
pyplot.show()
```
