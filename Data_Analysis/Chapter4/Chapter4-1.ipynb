# 任务4.1　读/写不同数据源的数据

## 代码4-1　SQLAlchemy连接MySQL数据库的代码

```python
from sqlalchemy import create_engine
engine = create_engine('mssql+pymssql://sa:123456@localhost:1433/testdb?charset=utf8')
print(engine)
```

## 代码4-2　使用`read_sql_table`、`read_sql_query`、`read_sql`函数读取数据库数据

```python
import pandas
formlist = pandas.read_sql_query('SELECT * FROM [sys].[tables]', con=engine)
print('testdb数据库数据表清单为：')
formlist
```

```python
detail1 = pandas.read_sql_table('meal_order_detail1', con=engine)
print('使用read_sql_table读取订单详情表的长度为：', len(detail1))
```

```python
detail2 = pandas.read_sql('SELECT * FROM [dbo].[meal_order_detail2]', con=engine)
print('使用read_sql函数+SQL语句读取的订单详情表长度为：', len(detail2))
detail3 = pandas.read_sql('meal_order_detail3', con=engine)
print('使用read_sql函数+表格名称读取的订单详情表长度为：', len(detail3))
```

## 代码4-3　使用`to_sql`方法写入数据

```python
detail1.to_sql('test1', con=engine, index=False, if_exists='replace')
formlist1 = pandas.read_sql_query('SELECT * FROM [sys].[tables]', con=engine)
print('新增一个表格后，testdb数据库数据表清单为：')
formlist1
```
