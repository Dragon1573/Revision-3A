## 实训1　读取并查看P2P网络贷款数据主表的基本信息

1. 训练要点
    1. 掌握`CSV`数据读取方法。
    2. 掌握`DataFrame`的常用属性与方法。
    3. 掌握`pandas`描述性统计方法。


2. 需求说明
    
    &emsp;&emsp;P2P网络贷款主表数据主要存放了网贷用户的基本信息。探索数据的基本信息，能够洞察数据的整体分布、数据的类属关系，从而发现数据间的关联。


3. 实现思路及步骤
    1. 使用`ndim`、`shape`、`memory_usage`属性分别查看维度、大小、占用内存信息。
    2. 使用`describe`方法进行描述性统计，并剔除值相同或全为空的列。

```python
import numpy
import pandas
table = pandas.read_csv('data/Training_Master.csv', sep=',', encoding='gbk')
print('主表的维度为：', table.ndim, '\n')
print('主表的大小为：', table.shape, '\n')
print('主表占用的内存空间为：', table.memory_usage(), sep='\n')
```

```python
print('主表的统计性描述为：')
table.describe()
```

```python
def drop_None_Columns(frame: pandas.DataFrame):
    before_clean = frame.shape[1]
    # 如果一列的数据全为空（NaN），则其计数为0
    column_is_None = (frame.describe().loc['count'] == 0)
    for column in range(len(column_is_None)):
        if column_is_None[column]:
            frame.drop(column_is_None.index[column], axis=1, inplace=True)
    # 如果一列的数据完全相同，则其标准差为0
    all_same_columns = (frame.describe().loc['std'] == 0)
    for column in range(len(all_same_columns)):
        if all_same_columns[column]:
            frame.drop(all_same_columns.index[column], axis=1, inplace=True)
    after_clean = frame.shape[1]
    print('源数据有', before_clean, '列')
    print('已清理', (before_clean - after_clean), '列')
    print('剩余', after_clean, '列有效')
drop_None_Columns(table)
```

## 实训2　提取用户信息更新表和登录信息表的时间信息

1. 训练要点
    1. 掌握时间字符串和标准时间的转换方法。
    2. 掌握时间信息提取的方法。

2. 需求说明

    &emsp;&emsp;用户信息更新表和登录信息表中均存在大量的时间数据，提取时间数据内存在的信息，一方面可以加深对数据的理解，另一方面能够探索这部分信息和目标的关联程度。同时用户登陆时间、借款成交时间、用户信息更新时间这些时间的时间差页能够反映出P2P网络贷款不同用户的行为信息。


3. 实现思路及步骤
    1. 使用`to_datetime`函数转换童虎信息更新表和登录信息表的时间字符串。

    3. 计算用户信息更新表和登录信息表中两时间的差，分别以日、小时、

```python
# 处理用户信息更新表
update_table = pandas.read_csv('data/Training_Userupdate.csv', sep=',', encoding='gbk')
update_table['ListingInfo1'] = pandas.to_datetime(update_table['ListingInfo1'])
update_table['UserupdateInfo2'] = pandas.to_datetime(update_table['UserupdateInfo2'])
print('转换后的用户信息更新表前5行数据为：', update_table.head(), '', sep='\n')
# 处理登录信息表
login_table = pandas.read_csv('data/Training_LogInfo.csv', sep=',', encoding='gbk')
login_table['Listinginfo1'] = pandas.to_datetime(login_table['Listinginfo1'])
login_table['LogInfo3'] = pandas.to_datetime(login_table['LogInfo3'])
print('转换后的登录信息表前5行数据为：', login_table.head(), sep='\n')
```

```python

update_time = pandas.DataFrame({
    'Listing_year': [row.year for row in update_table['ListingInfo1']],
    'Listing_month': [row.month for row in update_table['ListingInfo1']],
    'Listing_day': [row.day for row in update_table['ListingInfo1']],
    'Listing_week': [row.week for row in update_table['ListingInfo1']],
    'Listing_dayname': [row.day_name() for row in update_table['ListingInfo1']],
    'Update_year': [row.year for row in update_table['UserupdateInfo2']],
    'Update_month': [row.month for row in update_table['UserupdateInfo2']],
    'Update_day': [row.day for row in update_table['UserupdateInfo2']],
    'Update_week': [row.week for row in update_table['UserupdateInfo2']],
    'Update_dayname': [row.day_name() for row in update_table['UserupdateInfo2']]
})
print('用户信息更新表时间信息前5行为：')
update_time.head()
```

```python
login_time = pandas.DataFrame({
    'Listing_year': [row.year for row in login_table['Listinginfo1']],
    'Listing_month': [row.month for row in login_table['Listinginfo1']],
    'Listing_day': [row.day for row in login_table['Listinginfo1']],
    'Listing_week': [row.week for row in login_table['Listinginfo1']],
    'Listing_dayname': [row.day_name() for row in login_table['Listinginfo1']],
    'Login_year': [row.year for row in login_table['LogInfo3']],
    'Login_month': [row.month for row in login_table['LogInfo3']],
    'Login_day': [row.day for row in login_table['LogInfo3']],
    'Login_week': [row.week for row in login_table['LogInfo3']],
    'Login_dayname': [row.day_name() for row in login_table['LogInfo3']]
})
print('登录信息表时间信息前5行为：')
login_time.head()
```

## 实训3　使用分组聚合方法进一步分析用户信息更新表和登录信息表

1. 训练要点
    1. 掌握分组聚合的原理与步骤。
    2. 掌握`agg`、`apply`聚合方法。
    3. 掌握`transform`聚合方法。


2. 需求说明

    &emsp;&emsp;分析用户信息更新表和登录信息表时，除了提取时间本身的信息外，还可以结合用户编号进行分组聚合，然后进行组内分析。通过组内分析可以得出每组组内的最早和最晚信息更新时间、最早和最晚登录时间、信息更新的次数、登录的次数等信息。


3. 实现思路及步骤
    1. 使用`groupby`方法对用户信息更新表和登录信息表进行分组。
    2. 使用`agg`方法求取分组后的最早和最晚更新及登录时间。
    3. 使用`size`方法求取分组后的数据的信息更新次数与登录次数。

```python
update_group = update_table[['Idx', 'UserupdateInfo2']].groupby(by='Idx')
login_group = login_table[['Idx', 'LogInfo3']].groupby(by='Idx')
```

```python
print(
    '用户信息更新表按用户编号分组后，每组最早和最晚更新时间为：',
    update_group['UserupdateInfo2'].agg([numpy.min, numpy.max]),
    '', sep='\n'
)
print(
    '登录信息表按用户编号分组后，每组最早和最晚登录时间为：',
    login_group['LogInfo3'].agg([numpy.min, numpy.max]),
    sep='\n'
)
```

```python
print(
    '用户信息更新表按用户编号分组后，每组的信息更新次数前5行为：',
    update_group.count().head(),
    '', sep='\n'
)
print(
    '登录信息表按用户编号分组后，每组的登录次数前5行为：',
    login_group.count().head(),
    sep='\n'
)
```
