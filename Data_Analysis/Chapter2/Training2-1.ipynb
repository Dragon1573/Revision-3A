# 实训1　创建数组并进行运算

## 实验任务及其说明

1. 训练要点

    1. 掌握`NumPy`的数组创建及随机数生成。
    2. 掌握`NumPy`中用于统计分析的基本运算函数。

2. 需求说明

    `NumPy`数组在数值运算方面的效率优于`Python`提供的`list`，所以灵活掌握`NumPy`中数组的创建以及基础的运算时有必要的。
    
3. 实现思路及步骤

    1. 创建一个数值范围为$0$～$1$，间隔为$0.01$的数组。
    2. 创建100个副总正态分布的随机数。
    3. 对创建的2个数组进行四则运算。
    4. 对创建的随机数组进行简单的统计分析。

## 实验源代码

```python
# 导入 NumPy 程序包
import numpy
```

```python
# 生成数组，共100项元素
avg_rand = numpy.arange(0, 1, 0.01)
print('均匀分布随机数组：', avg_rand, sep='\n')
```

```python
# 以学号作为随机数种子，确保随机结果可复现
numpy.random.seed(60123)
# 生成100项服从正态分布 N(5, 2) 的随机数
norm_rand = 2 * numpy.random.randn(100) + 5
print('满足N(5, 2)的正态分布随机数组：', norm_rand, sep='\n')
```

```python
# 进行四则运算
print('数组加法：', avg_rand + norm_rand, sep='\n')
```

```python
print('数组减法（大数减小数）：', norm_rand - avg_rand, sep='\n')
```

```python
print('数组乘法（对应项乘法）：', avg_rand * norm_rand, sep='\n')
```

```python
print(
    '数组除法（对应项除法，以大除小防止分母为0）：',
    avg_rand / norm_rand, sep='\n'
)
```

```python
# 以正态随机数组为分析目标
print('算术平均数为：', norm_rand.mean())
```

```python
print('总体标准差为：', norm_rand.std())
```

```python
print('总体方差为：', norm_rand.var())
```

```python
print('最小值为：', norm_rand.min())
```

```python
print('最大值为：', norm_rand.max())
```

```python
print('元素求和为：', norm_rand.sum())
```

```python
print('元素求积为：', norm_rand.prod())
```

```python
print('最小值索引为：', norm_rand.argmin())
```

```python
print('最大值索引为：', norm_rand.argmax())
```

```python
print('升序排序结果为：', numpy.sort(norm_rand), sep='\n')
```
